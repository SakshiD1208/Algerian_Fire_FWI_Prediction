```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()
```

```python
df.info()
#Convert numerical values in int or float
```

**DATA CLEANING**


```python
#Check for null values
df.isnull().sum()
```

```python
#Which colums are null
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
#Create sepearte col for region and for Bejaia region assign 0 and for sidi-Bel abbes region assign 1
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.head()
```

```python
df.info()
```

```python
#Covert region float to in
df[['Region']]=df[['Region']].astype(int)
df.head()
```

```python
#Remove Null value
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
#Drop 122 row as it is just header
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns
#There is a extra space betwwen col and we need to fix that
```

```python
#Fixing extra space in col
df.columns=df.columns.str.strip()
df.columns
```

```python
#Change Required col datatype to in
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Save the file
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```

**EXPLORATORY DATA ANALYSIS**

```python
#drop unnecessay feature
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
## Encoding of the categories in classes
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
import matplotlib.pyplot as plt

# Set the style for the plots
plt.style.use('seaborn-v0_8')  # Updated style to avoid deprecation warnings

# Plot histograms for all numeric features in the DataFrame
df_copy.hist(bins=50, figsize=(20, 15), edgecolor='black')

# Display the plots
plt.tight_layout()  # Adjusts subplots to fit into the figure area.
plt.show()
```

```python
#Percentage of fire and no fire region

percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#Pie chart for visulaization of precentage
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
